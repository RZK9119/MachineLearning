```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from ipywidgets import interact
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = '/content/drive/MyDrive/Dataset/diamonds.csv'

df = pd.read_csv(data)

df.head()
```

```python
features = ['carat', 'cut', 'color', 'clarity', 'depth', 'table', 'price', 'x', 'y', 'z']

print("unique value untuk masing-masing fitur:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()}")
```

```python
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']
categorical_features = ['cut', 'color', 'clarity']
target_feature = ['price']
```

```python
for feature in categorical_features:
    data_counts = df[feature].value_counts()
    plt.figure(figsize=(8,6))
    data_counts.plot(kind='bar', color='violet')
    plt.title(f'Bar Chart of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Counts')
    plt.show()
```

```python
for feature in categorical_features:
    data_counts = df[feature].value_counts()
    plt.figure(figsize=(8,6))
    colors = plt.cm.tab10.colors
    plt.pie(data_counts, labels=data_counts.index, autopct='%1.1f%%', colors=colors)
    plt.title(f'Pie Chart of {feature}')
    plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, feature in enumerate(numerical_features):
    sns.histplot(df[feature], kde=True, color='violet', ax=axes[i])
    axes[i].set_title(f'Histogram of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
def plot_histogram(feature, bin_width):
    plt.figure(figsize=(8,6))
    sns.histplot(df[feature], kde=True, bins=np.arange(min(df[feature]), max(df[feature]) + bin_width, bin_width), color='violet')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()

for feature in numerical_features:
    bin_slider = widgets.FloatSlider(min=0.1, max=1.0, step=0.1, value=0.5, description="Bin Width")
    interactive_plot = widgets.interactive(plot_histogram, feature=widgets.fixed(feature), bin_width=bin_slider)
    display(interactive_plot)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, feature in enumerate(numerical_features):
    sns.boxplot(x=df[feature], ax=axes[i], color='violet')
    axes[i].set_title(f'Box Plot of {feature}')
    axes[i].set_xlabel(feature)

plt.tight_layout()
plt.show()
```
