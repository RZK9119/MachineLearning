```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from ipywidgets import interact
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = '/content/drive/MyDrive/Dataset/kc_house_data.csv'

df = pd.read_csv(data)

df.head()
```

```python
categorical_features = ['date', 'waterfront', 'view', 'condition', 'grade', 'yr_built', 'yr_renovated', 'zipcode']
numerical_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'sqft_above', 'sqft_basement', 'lat', 'long', 'sqft_living15', 'sqft_lot15']
target = 'price'
```

```python
for feature in categorical_features:
    plt.figure(figsize=(8, 6))
    plt.hist(df[feature], bins=20, color='Violet')
    plt.title(f'Distribusi {feature}', fontsize=15)
    plt.xlabel(feature, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(axis='y', alpha=0.75)
    plt.show()
```

```python
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df, x=feature, kde=True, bins=20, color='Violet')
    plt.title(f'Distribusi {feature}', fontsize=15)
    plt.xlabel(feature, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(axis='y', alpha=0.75)
    plt.show()
```

```python
fig, axes = plt.subplots(4, 3, figsize=(15, 10))
axes = axes.flatten()

for i, feature in enumerate(numerical_features):
    sns.histplot(df[feature], kde=True, color='violet', ax=axes[i])
    axes[i].set_title(f'Histogram of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')

for i in range(len(numerical_features), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

```python
def plot_histogram(feature, bin_width):
    plt.figure(figsize=(8,6))
    sns.histplot(df[feature], kde=True, bins=np.arange(min(df[feature]), max(df[feature]) + bin_width, bin_width), color='violet')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()

for feature in numerical_features:
    bin_slider = widgets.FloatSlider(min=1, max=1000, step=1, value=20, description="Bin Width")
    interactive_plot = widgets.interactive(plot_histogram, feature=widgets.fixed(feature), bin_width=bin_slider)
    display(interactive_plot)
```

```python
fig, axes = plt.subplots(4, 3, figsize=(15, 10))
axes = axes.flatten()

for i, feature in enumerate(numerical_features):
    sns.boxplot(x=df[feature], ax=axes[i], color='violet')
    axes[i].set_title(f'Box Plot of {feature}')
    axes[i].set_xlabel(feature)

for i in range(len(numerical_features), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```
